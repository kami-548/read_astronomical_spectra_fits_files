```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

# read fits file spectra
file = fits.open('arp_105.fits')

# for file information
file.info()

```

```python
# reading flux, wavelength and redshift
data = file[1].data
flux = data['flux']
wave = 10**data['loglam']

# for ploting
plt.plot(wave, flux)
plt.title('Galaxy Spectra of ARP 105 in moving frame')
plt.xlabel('Wavelength $\AA$')
plt.ylabel('Flux [10^{-17}] erg/cm^2/s/$\\AA$')
```

There are some emissions lines and some absorption lines aswell

```python
# for the value of the redshift
z = file[2].data['z']
z[0]
```

So the redshift is 0.02874983.

## Now Spectrum for the Rest Frame for z = 0.028749827

```python
plt.figure()
plt.plot(wave/(1 + z), flux, color= 'red', label = 'Spectra')
plt.title('Galaxy Spectra of ARP 105 in Rest frame')
plt.xlabel('Rest Wavelength $\AA$')
plt.ylabel('Flux [10^{-17}] erg/cm^2/s/$\\AA$')
plt.axvline(x= 6564.16, label= 'H-alpha line')
plt.axvline(x = 3729.8, label = 'OII', color = 'black')
plt.legend()
```

## for Comparison

```python
plt.plot(wave, flux)
plt.plot(wave/(1 + z), flux, color= 'red')
```

we can see that the spectra is little shifted.
